```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
image = cv2.imread('img/295257304_de893fc94d.jpg') 
```

```python
# Convert the image to different color spaces
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
# Display the images using Matplotlib
plt.figure(figsize=(12, 8))
```

```python
# Original Image (BGR)
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image (BGR)')
# plt.axis('off')
```

```python
# RGB Image
plt.subplot(2, 2, 2)
plt.imshow(image_rgb)
plt.title('RGB Image')
# plt.axis('off')
```

```python
# Grayscale Image
plt.subplot(2, 2, 3)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
# plt.axis('off')
```

```python
# HSV Image
plt.subplot(2, 2, 4)
plt.imshow(image_hsv)
plt.title('HSV Image')
# plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('img/295257304_de893fc94d.jpg')

# Convert the image to float
image_float = image.astype(float)

# Define the luminosity weights
weights = np.array([0.299, 0.587, 0.114])

# Apply the luminosity formula
grayscale_image = np.dot(image_float[..., :3], weights)

# Convert the grayscale image to uint8
grayscale_image = np.clip(grayscale_image, 0, 255).astype(np.uint8)

# Display the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.title('Grayscale Image (Luminosity)')
plt.axis('off')
plt.show()
```
